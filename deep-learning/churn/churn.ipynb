**Churn prediction** 


This code predicts the likelihood of a customer churning. Data is from a service provider 

Import Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Churn.csv')
df.head(5)
```

```python
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1)) #hot encoding - get_dummies encodes text to numbers
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0) #lambda?
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

Import Dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

Build and Compile model

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd')
```

Fit, Predict and Evaluate

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_pred = model.predict(X_test)
y_pred = [0 if val < 0.5 else 1 for val in y_pred]
```

```python
accuracy_score(y_test, y_pred)
```
